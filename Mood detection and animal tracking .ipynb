# Elephant

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile

# Read the audio file
frequency_sampling, audio_signal = wavfile.read("D:/AI CCP/Animal Sounds/Elephant/elephant4.wav")

# Normalize the audio signal
if audio_signal.dtype == np.int16:
    audio_signal = audio_signal / np.power(2, 15)
elif audio_signal.dtype == np.int32:
    audio_signal = audio_signal / np.power(2, 31)

# Remove the first 100 samples
audio_signal = audio_signal[100:]

# Calculate the maximum amplitude
max_amplitude = np.max(np.abs(audio_signal))

# Classify the mood based on the amplitude
if max_amplitude >= 0.70:
    mood = "Angry"
elif 0.50 <= max_amplitude < 0.70:
    mood = "Normal/Happy"
else:
    mood = "Sad" 

print('\nSignal shape:', audio_signal.shape)
print('Signal Datatype:', audio_signal.dtype)
print('Signal duration:', round(audio_signal.shape[0] / float(frequency_sampling), 2), 'seconds')
print(f"Maximum Amplitude: {max_amplitude:.2f}")
print(f"Classified Mood of Elephant: {mood}")

# Create a time axis in milliseconds for the modified audio signal
time_axis = 1000 * np.arange(0, len(audio_signal), 1) / float(frequency_sampling)

# Plot the modified audio signal
plt.figure(figsize=(12, 6))
plt.plot(time_axis, audio_signal, color='purple')
plt.xlabel('Time (milliseconds)')
plt.ylabel('Amplitude')
plt.title(f'Classified Mood of Elephant: {mood}')
plt.show()
```

# BFS

```python
import networkx as nx
import matplotlib.pyplot as plt

# Define the graph
G = nx.Graph()

# Add nodes for each location
locations = [
    'Savannah', 'Forest', 'Riverbank', 'Mountain', 'Grassland', 'Desert', 'Lake', 'Hills', 'Valley', 'Waterfall',
    'Plains', 'Oasis', 'Plateau', 'Cave'
]
G.add_nodes_from(locations)

# Add edges to represent possible paths (connections between locations)
edges = [
    ('Savannah', 'Forest'), ('Savannah', 'Riverbank'), ('Forest', 'Riverbank'),
    ('Forest', 'Mountain'), ('Riverbank', 'Grassland'), ('Mountain', 'Grassland'),
    ('Desert', 'Oasis'), ('Lake', 'Riverbank'), ('Hills', 'Valley'), ('Plains', 'Grassland'), ('Oasis', 'Desert'),
    ('Plateau', 'Hills'), ('Waterfall', 'Riverbank'), ('Mountain', 'Cave'),
    ('Plains', 'Desert'), ('Mountain', 'Desert'), ('Hills', 'Lake'), ('Valley', 'Waterfall')
]
G.add_edges_from(edges)

# Define starting positions and destinations for animals
animal_positions = {
    'lion1': ('Savannah', 'Forest'),
    'lion2': ('Mountain', 'Cave'),
    'lion3': ('Grassland', 'Lake'),
    'lion4': ('Plains', 'Desert'),
    'lion5': ('Forest', 'Riverbank'),
    'lion6': ('Forest', 'Valley'),
    'lion7': ('Riverbank', 'Waterfall'),
    'fox1': ('Forest', 'Cave'),
    'fox2': ('Waterfall', 'Riverbank'),
    'fox3': ('Cave', 'Mountain'),
    'fox4': ('Mountain', 'Savannah'),
    'fox5': ('Hills', 'Lake'),
    'elephant1': ('Riverbank', 'Desert'),
    'elephant2': ('Riverbank', 'Grassland'),
    'elephant3': ('Lake', 'Hills'),
    'elephant4': ('Forest', 'Valley'),
    'elephant5': ('Savannah', 'Oasis'),
    'elephant6': ('Desert', 'Plateau'),
    'elephant7': ('Oasis', 'Cave'),
    'horse1': ('Mountain', 'Plains'),
    'horse2': ('Plateau', 'Savannah'),
    'horse3': ('Valley', 'Forest'),
    'horse4': ('Plateau', 'Desert'),
    'horse5': ('Plains', 'Riverbank')
}

# Function to simulate animal movement using BFS
def simulate_animal_movement(graph, animal_positions):
    for animal, (start_position, end_position) in animal_positions.items():
        print(f"{animal} starting at {start_position}, aiming to reach {end_position}")
        visited = set()  # Set to keep track of visited nodes
        queue = [(start_position, [start_position])]  # Queue initialized with the starting position and the path taken so far
        found = False
        
        while queue and not found:
            (current_node, path) = queue.pop(0)  # Dequeue the next node to process
            if current_node == end_position:
                print(f"{animal} reached its destination: {end_position}")
                found = True  # End search when destination is reached
            if current_node not in visited:
                visited.add(current_node)  # Mark the current node as visited
                neighbors = list(graph.neighbors(current_node))  # Get the neighbors of the current node
                for neighbor in neighbors:
                    if neighbor not in visited:
                        queue.append((neighbor, path + [neighbor]))  # Enqueue unvisited neighbors with updated path
        print(f"{animal} finished its path: {' -> '.join(path)}\n")

# Simulate animal movement
simulate_animal_movement(G, animal_positions)

# Visualize the graph
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color='lightgreen', node_size=2000, font_size=10, font_weight='bold', edge_color='gray')
nx.draw_networkx_edges(G, pos, edge_color='blue')
plt.title('Graph-Based Movement Tracking of Animals')
plt.show()
```
